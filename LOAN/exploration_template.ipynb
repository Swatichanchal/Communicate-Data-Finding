#### - The function which is going to help throughtout this project , as we have to mention X - lable , y-lable , title each and every time we plot any graph for clear understanding .

```python
def lables_name(x , y = None):
    plt.xlabel(x)
    plt.ylabel(y)
    if(y == None):
        plt.title(x , 'Distribution');
    else:
        plt.title(f' {x} and {y} Distribution')
```

```python
# convert LoanOriginationDate column to datetime
df['LoanOriginationDate']= pd.to_datetime(df['LoanOriginationDate'])
```

```python
# Extract year, month, day, hour, and minute. Assign all these date component to new column.
df['Originationyear'] = df['LoanOriginationDate'].dt.year
df['Originationmonth'] = df['LoanOriginationDate'].dt.month
df['Originationday'] = df['LoanOriginationDate'].dt.day

df.head(2)
```
